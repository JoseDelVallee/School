```python
from numpy import *
import matplotlib.pyplot as plt

fig,ax = plt.subplots(2,3,figsize=(12,12))# Crea una figura con 3 subfiguras


filas = 250
columnas = 250

#Matriz A Y B

A = zeros((filas,columnas))
B = zeros((filas,columnas))

#For de la matriz A
for i in range(filas):
    for j in range(columnas):
        A[i][j] = i

#For de la matriz B de izquierda a Derecha
for i in range(filas):
    for j in range(columnas):
        B[i][j] = j
producto = dot(A,B)

print(A)
print(B)
print(producto)

ax[0, 0].imshow(A, cmap='Reds')
ax[0, 0].set_title('Matriz A')

ax[0,1].imshow(B, cmap='Reds')
ax[0,1].set_title('Matriz B')

ax[0,2].imshow(producto, cmap='Reds')
ax[0,2].set_title('Producto A*B')

# Tamaño del círculo y centro
porcentaje = (filas/2) * 0.4    #Porcentaje se calcula como el 40% del número de filas dividido por 2.
#Esto define un porcentaje del tamaño delas(filas) que se utilizará como radio del círculo.
radio = porcentaje  # Radio del círculo
centro_x = (int)(filas/2)  # Coordenada x del centro del círculo
centro_y = (int)(columnas/2)  # Coordenada y del centro del círculo

C = zeros((filas, columnas))
D = zeros((filas, columnas))

for i in range(filas):
    for j in range(columnas):
        # Calcula la distancia desde el centro del círculo
        distancia = sqrt((i - centro_x) ** 2 + (j - centro_y) ** 2)
        # Asigna el valor en función de la distancia en A
        C[i][j] = distancia

# Matriz B de mayor a menor
for i in range(filas):
    for j in range(columnas):
        # Calcula la distancia desde el centro del círculo
        distancia = sqrt((i - centro_x) ** 2 + (j - centro_y) ** 2)
        # Asigna el valor en función de la distancia en B (invierte la distancia)
        D[i][j] = radio - distancia

ax[1,0].imshow(C, cmap='Blues')  # Visualización de la matriz A en la subfigura
ax[1,0].set_title('Matriz C')  # Título de la subfigura

ax[1,1].imshow(D, cmap='Blues')
ax[1,1].set_title('Matriz D')

# Producto matricial
E = dot(C, D)
ax[1,2].imshow(E, cmap='Blues')
ax[1,2].set_title('Producto C*D')

plt.show()
```
